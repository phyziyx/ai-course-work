```python
# 1. Write a programme that creates a person dictionary with the following keys: RegNo, Nmae, Number, Email, Uni, Year.
# Take input for all keys and save in a file called regno.txt

keyNames = ['RegNo', 'Name', 'Number', 'Email', 'Uni', 'Year']
person = {}

for key in keyNames:
    value = input(f'Please insert {key}')
    person[key] = value

file = open(f'{person['RegNo']}.txt', 'w')
file.write(str(person))
file.close()
```

```python
from collections import defaultdict

# edges = [['A', 'B'], ['A', 'C'], ['B', 'D'], ['B', 'E'], ['C', 'F'], ['C', 'G']]
edges = [['2', '4'], ['2', '6'], ['4', '8'], ['4', '10'], ['6', '12'], [
    '6', '14'], ['8', '16'], ['8', '18'], ['10', '20'], ['10', '22'], ['12', '24']]


def createAdjancenyList(edges):
    graph = defaultdict(list)
    for u, v in edges:
        graph[u].append(v)
        graph[v].append(u)
    return graph


def bfs(graph, start):
    visitedSet = set()
    visitedSet.add(start)

    queue = []
    queue.append(start)
    result = []

    while (queue):
        current = queue[0]
        result.append(current)
        queue = queue[1:]
        for n in graph[current]:
            if n not in visitedSet:
                visitedSet.add(n)
                queue.append(n)
    return result


adjacentList = createAdjancenyList(edges)
print(bfs(adjacentList, '2'))


def dfs(graph, start, visitedSet=None, path=None):
    if visitedSet is None:
        visitedSet = set()
    if path is None:
        path = []

    visitedSet.add(start)
    path.append(start)
    if start in graph:
        for n in graph[start]:
            if n not in visitedSet:
                dfs(graph, n, visitedSet, path)
    return path


print(dfs(adjacentList, '2'))
```
